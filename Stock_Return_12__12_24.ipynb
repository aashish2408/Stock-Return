## Firstly i used pandas library to import csv file.

```python
import pandas as pd
df = pd.read_csv("Stock_data.csv")
```

```python
print(df)
```

## Then make a list of the independent values and call this variable x and then put the dependent values in a variable called y.

```python
x = df[['Volatility', 'Market Return']]
y = df['Stock return']
```

## I will use some methods from the sklearn module and from the sklearn module we will use the LinearRegression() method to create a linear regression. Then i will fit the independent and dependent values by using fit()

```python
from sklearn import linear_model

regr = linear_model.LinearRegression()
regr.fit(x, y)
```

## Calculating coefficients

```python
print(regr.coef_)
```

## Calculating Intercept

```python
print(regr.intercept_)
```

## Predicting stock return when by asumming market return is 0.08 and volatility is 0.31.

```python
prediction = regr.predict([[0.31, 0.08]])

print(prediction)
```

# Therefore possible linear equation is : Stock return = 0.012*(Volatility) + 1.2*(Market return) + 0.04
